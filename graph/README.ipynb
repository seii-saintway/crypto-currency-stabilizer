```python
# Launch Classic Notebook
# 打开matplotlib的交互显示模式
%matplotlib notebook
import matplotlib.pyplot as plt
import time

plt.figure()
# 绘制一条空曲线，令下面ax.lines里不为空
plt.plot([], [])

# 获取当前figure和axis
fig = plt.gcf()
ax = plt.gca()

for i in range(10):
    # 正常打印
    print(i, end=' ')

    x = list(range(i, i+10))
    y = [xx**2 for xx in x]

    # 更新数据
    ax.lines[0].set_data(x, y)

    # 重置已有曲线的的坐标轴时，要先解除限制（见ax.autoscale_view的官方文档）
    ax.relim()
    # 重设坐标轴，让图像更协调
    ax.autoscale_view()

    # 显示图片
    fig.canvas.draw()

    time.sleep(0.1 * 10)
```

```python
# %matplotlib inline
import matplotlib.pyplot as plt
plt.close('all')

import time
import IPython

ax = []                    # 定义一个 x 轴的空列表用来接收动态的数据
ay = []                    # 定义一个 y 轴的空列表用来接收动态的数据
plt.ion()                  # 开启一个画图的窗口
for i in range(100):       # 遍历0-99的值
    ax.append(i)           # 添加 i 到 x 轴的数据中
    ay.append(i**2)        # 添加 i 的平方到 y 轴的数据中
    # plt.clf()              # 清除之前画的图
    IPython.display.display(plt.gcf())
    IPython.display.clear_output(wait=True)
    plt.plot(ax, ay)        # 画出当前 ax 列表和 ay 列表中的值的图形
    # plt.pause(0.1)         # 暂停
    time.sleep(0.1)
    plt.ioff()             # 关闭画图的窗口
```

```python
# %matplotlib inline
import matplotlib.pyplot as plt
plt.close('all')

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from mpl_toolkits.mplot3d import Axes3D
import time
import IPython

# 解决中文乱码问题
# myfont = fm.FontProperties(fname="simsun.ttc", size=14)
matplotlib.rcParams["axes.unicode_minus"] = False


def simple_plot():
    """
    simple plot
    """
    # 生成画布
    plt.figure(figsize=(8, 6), dpi=80)

    # 打开交互模式
    plt.ion()

    # 循环
    for index in range(20):
        # 清除原有图像
        # plt.cla()
        IPython.display.display(plt.gcf())
        IPython.display.clear_output(wait=True)

        # 设定标题等
        plt.title("Dynamic Curve")#, fontproperties=myfont)
        # 网格线
        plt.grid(True)

        # 生成测试数据
        x = np.linspace(-np.pi + 0.1*index, np.pi+0.1*index, 256, endpoint=True)
        y_cos, y_sin = np.cos(x), np.sin(x)

        # 设置X轴
        plt.xlabel("X axis")#, fontproperties=myfont)
        plt.xlim(-4 + 0.1*index, 4 + 0.1*index)
        plt.xticks(np.linspace(-4 + 0.1*index, 4+0.1*index, 9, endpoint=True))

        # 设置Y轴
        plt.ylabel("Y axis")#, fontproperties=myfont)
        plt.ylim(-1.0, 1.0)
        plt.yticks(np.linspace(-1, 1, 9, endpoint=True))

        # 画两条曲线
        plt.plot(x, y_cos, "b--", linewidth=2.0)#, label='cos')
        plt.plot(x, y_sin, "g-", linewidth=2.0)#, label='sin')

        # 设置图例位置,loc可以为[upper, lower, left, right, center]
        # plt.legend(loc="upper left", shadow=True)#, prop=myfont)

        # 暂停
        # plt.pause(0.1)
        time.sleep(0.1)

    # 关闭交互模式
    plt.ioff()

    # 图形显示
    plt.show()


simple_plot()
```
